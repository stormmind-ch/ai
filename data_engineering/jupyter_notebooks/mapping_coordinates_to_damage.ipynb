```python
import pandas as pd
```

```python
crd_csv = "../../Ressources/municipalities_coordinates_newest.csv"
dmg_exl = "../../StormMindData/USDB_1972_2023_ohneSchadenszahlen_ohneBeschriebe.xlsx"
dmg_crd_csv = "../../Ressources/damage_coordinates.csv"
```

```python
crd = pd.read_csv(crd_csv, encoding="utf-8")
crd
```

```python
dmg = pd.read_excel(dmg_exl,skiprows=3)
dmg
```

```python
import csv
import pandas as pd
for _, row in dmg.iterrows():
    gemeinde = row["Gemeinde"]
    
    match = crd[crd["Municipality"] == gemeinde]

    if not match.empty:
        lat = match.iloc[0]["Latitude"]
        lng = match.iloc[0]["Longitude"]

        xxx = {
            "lat": lat,
            "lng": lng,
            "gemeinde": gemeinde,
            "datum": row["Datum"].strftime('%Y-%m-%d') #pd.to_datetime(dmg["Datum"])
        }

        with open(dmg_crd_csv, mode='a', newline='', encoding='utf-8') as f:
            writer = csv.DictWriter(f, fieldnames=["lat", "lng", "gemeinde", "datum"])
            writer.writerow(xxx)  # oder was du eben damit machen willst
```
