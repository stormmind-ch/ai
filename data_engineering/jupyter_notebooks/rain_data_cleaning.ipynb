```python
import glob

from fontTools.afmLib import kernRE
from pygments.lexers.csound import newline
```

```python
ordner_pfad = "../Ressourcen/Niederschlagsdaten"
datei_muster = f"{ordner_pfad}/*legend.txt"
zeilen_nummer = 32
kuerzel = []
```

```python
line = []
newline = []
for datei_pfad in glob.glob(datei_muster):
    try:
        with open(datei_pfad, "r", encoding='iso-8859-1') as f:
            zeilen = f.readlines()
            if len(zeilen) > zeilen_nummer:
                dings = zeilen[zeilen_nummer].split()
                inde = dings.index("rka150d0")
                kack = dings[:inde]
                kuerzel.append(kack[0] + "," + str(kack[1:]))
            else:
                print(f"Datei: {datei_pfad} hat weniger als {zeilen_nummer + 1} Zeilen.")
    except Exception as e:
        print(f"Fehler beim Lesen von {datei_pfad}: {e}")
```

```python

namen = kuerzel
print(namen)
```
